## Train test split

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X=df.drop("Selling_Price", axis=1)
y=df.Selling_Price
```

```python
cat = X.select_dtypes("object").columns
cat
```

```python
enc = OrdinalEncoder()
X[cat] = enc.fit_transform(X[cat])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=5)

print("Train features shape : ", X_train.shape)
print("Train target shape   : ", y_train.shape)
print("Test features shape  : ", X_test.shape)
print("Test target shape    : ", y_test.shape)
```

## Modeling

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {"train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    "test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

## Adaboost Regressor

AdaBoost, her bir aa iin arlkl tahminleri birletirerek nihai tahmini oluturur. lk birka aa genellikle dierlerinden daha byk bir arla sahip olabilir, ancak bu, AdaBoost'un yalnzca ilk aacn tahminini kulland anlamna gelmez. AdaBoost, tm aalarn tahminlerini birletirir, ancak her bir aacn arl, o aacn eitim veri seti zerindeki performansna gre belirlenir.

gelelim learning rate e bu rada 2. aata en sada olan 35 olan tahminime baknca 19 adet samples olduunu gryorum fiyat 35 olan ka tane ara vard datamda 1 adet. Fakat burann samples i 19 demekki 18 olarak artrd. imdi ben bunu learning rate ile oynarak dzelticem. gzlem saysn azaldka azaltyordu. Ama tekrar sylyorum learning rate ile oynarken mutlaka aa saysn da dengede tutmak lazm

#### loss

linear: Bu, hatalarn dorusal bir fonksiyon olarak kullanlmasn belirtir. Eer tahminler gerek deerlerden bykse, tahminleri azaltr; eer tahminler gerek deerlerden kkse, tahminleri artrr.

square: Hatalarn karesinin alnarak kullanld bir yntemdir. Bu, byk hatalarn daha byk bir etkisi olaca anlamna gelir.

exponential: Bu, AdaBoost'un klasik hata fonksiyonudur. Aslnda, AdaBoost'un orijinal algoritmasnda bu hata fonksiyonu kullanlr. Ancak, regresyon iin bu seenek genellikle kullanlmaz.

## Gridsearch for Adaboosting

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_estimators":[50, 100, 200, 300, 500],
              "learning_rate":[0.1, 0.5, 0.8, 1],
              "loss": ["linear", "square"]
            }
```

```python
model = AdaBoostRegressor(random_state=101)
grid_model = GridSearchCV(estimator=model,
                          param_grid=param_grid,
                          scoring='neg_root_mean_squared_error',
                          cv=10,
                          n_jobs = -1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
train_val(grid_model, X_train, y_train, X_test, y_test)
```

## Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
grad_model = GradientBoostingRegressor(random_state=101)
```

```python
# 100 -- 150 -- (-50) -- (-30) --> 150 + 0.1 * (-50) + 0.1*(-30)

# 200 -- 150 -- (50) -- (25)  --> 150 + 0.1 * 50 + 0.1*25
```

```python
grad_model.fit(X_train, y_train)
```

```python
train_val(grad_model, X_train, y_train, X_test, y_test)
```

```python
model = GradientBoostingRegressor(random_state=101)
scores = cross_validate(model, X_train, y_train, scoring=['r2', 
            'neg_mean_absolute_error','neg_mean_squared_error','neg_root_mean_squared_error'], cv =10)
pd.DataFrame(scores)
pd.DataFrame(scores).mean()[2:]
```
