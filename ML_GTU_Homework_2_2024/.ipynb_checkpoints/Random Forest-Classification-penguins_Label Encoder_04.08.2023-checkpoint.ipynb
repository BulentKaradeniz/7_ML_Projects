## Label Encoder 

Bu yntemde, kategorik deikenin her bir deeri benzersiz bir say ile eletirilir.

LabelEncoder: Tek bir stunu kodlamak iin kullanlr. Genellikle hedef deikenler (Target) iin idealdir.

## Ordinal Encoder

Sral kategorik deikenler iin kullanlr. rnein "dk", "orta", "yksek" deerlerini srasyla 0, 1, 2 olarak kodlayabilirsiniz.

OrdinalEncoder: Birden fazla kategorik stunu ayn anda kodlamak iin kullanlr. zellik stunlar iin daha uygundur.

Eer kategorik deikenler sral (ordinal) bir yapya sahipse, bu sralamay koruyarak kodlama yapmak nemlidir. Bu, modelin doru renme yapabilmesi iin gereklidir.

OrdinalEncoder ile bu tr sral kategorik veriyi dntrmek iin, sralamanz belirten bir liste kullanarak kodlama yapabilirsiniz. rnein "dk", "orta" ve "yksek" sralamasn koruyarak kodlamak isterseniz u ekilde bir yaklam izleyebilirsiniz:

```python
from sklearn.preprocessing import OrdinalEncoder

# rnek veri
data = [["orta"], ["dk"], ["yksek"], ["dk"], ["yksek"], ["orta"]]

# Sralamay belirtmek iin kategori listesi
categories = [['dk', 'orta', 'yksek']]

encoder = OrdinalEncoder(categories=categories)
encoded_data = encoder.fit_transform(data)

print(encoded_data)
```

Eer bir zellik iin kategorik deerlerin sras nemli deilse ve OrdinalEncoder' sralama belirtmeden kullanrsanz, bu durumda OrdinalEncoder'n yapt dnm, LabelEncoder'n tek bir stun iin yapaca dnme benzer ekilde alfabetik sralama zerinden gerekleir.

Ancak, u nemli farklar unutmamak gerekir:

**LabelEncoder**, yalnzca tek bir stunu dntrmek iin tasarlanmtr. **OrdinalEncoder** ise birden ok stunu ayn anda dntrebilir.

**LabelEncoder** genellikle hedef deikenleri (etiketleri) dntrmek iin nerilirken, **OrdinalEncoder** zellik stunlar (features) iin nerilir.

zetle, eer bir zelliin kategorik deerlerinin sras nemli deilse ve bu deerleri saysal formata dntrmek istiyorsanz, LabelEncoder veya OrdinalEncoder' (sralama belirtmeden) kullanabilirsiniz.

Ancak, **birden ok stunu ayn anda dntrme** ihtiyacnz varsa, bu durumda OrdinalEncoder daha uygun bir seenek olacaktr.

Uzun lafn ksas ordinal encoder kullanmak daha mantkl. Sadece hiyerari ieren features iin hiyerari oluturmak nemli!!

# Peki hangisini hangisinde kullanacaz?

### Lineer Regresyon Modelleri

**Nominal Deikenler:** One-Hot Encoding (get_dummies). Ancak multicollinearity sorununu engellemek iin bir kategori drmelisiniz (dummy variable trap'ten kanmak iin).

**Ordinal Deikenler:** Ordinal Encoding veya belirli bir sralama mant varsa map fonksiyonu.

### Aa Tabanl Modeller (Karar Aalar, Rastgele Ormanlar, XGBoost, LightGBM vs.)

**Nominal ve Ordinal Deikenler:** Bu modeller kategorik deikenlerle baa kabilecek kapasiteye sahip olduklar iin genellikle Label Encoding (ordinal encoding) yeterli olacaktr. zellikle LightGBM, kategorik deikenleri dorudan ileyebilme kapasitesine sahiptir. One-Hot Encoding, bu tr modellerde performans drebilir ve modelin eitim sresini uzatabilir.


### Lojistik Regresyon

**Nominal Deikenler:** One-Hot Encoding.

**Ordinal Deikenler:** Ordinal Encoding.


### Destek Vektr Makineleri (SVM), K-En Yakn Komu (KNN) gibi mesafe tabanl algoritmalar

**Nominal Deikenler:** One-Hot Encoding tercih edilmelidir, nk bu algoritmalar zellikler arasndaki mesafeye dayaldr ve nominal deikenler iin label encoding yanltc olabilir.

**Ordinal Deikenler:** Ordinal Encoding.


### Sinir Alar (Neural Networks)

**Nominal Deikenler:** One-Hot Encoding genellikle tercih edilir.

**Ordinal Deikenler:** Ordinal Encoding veya One-Hot Encoding kullanlabilir. Modelin karmaklna ve veri setinin yapsna bal olarak en iyi sonucu veren yaklam semek iin denemeler yaplabilir.


### Naive Bayes

**Nominal Deikenler:** One-Hot Encoding.

**Ordinal Deikenler:** Ordinal Encoding ya da Label Encoding.


Yukarda sadece literatr destekli ahsi nerilerimi yazdm. zellikle karmak ve byk veri setleri zerinde alrken, hangi kodlama ynteminin en iyi sonucu verdiini grmek iin best practice kavram aklnzdan kmasn. Bu kapsamda denemeler yapmak ve apraz dorulama(CV) yntemleri kullanarak model performansn deerlendirmek nemlidir. Ksaca en iyi skor hangisinde ise oradan devam etmek doru olur.

## Modelling and Model Performance

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix

def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
dt_model = DecisionTreeClassifier(random_state=101)
dt_model.fit(X_train, y_train)
eval_metric(dt_model, X_train, y_train, X_test, y_test)

# encoder yaptmzda sonular daha iyi kt  
```

### Decision Tree

```python
dt_model = DecisionTreeClassifier(class_weight = "balanced", random_state=101)
```

```python
dt_model.fit(X_train, y_train)
```

__Cross Validate__

```python
model = DecisionTreeClassifier(class_weight = "balanced", random_state=101)

scores = cross_validate(model, X_train, y_train, scoring = ['accuracy', 'precision', 'recall', 'f1'], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))

df_scores.mean()[2:]
```

__GridSearch__

```python
param_grid = {"splitter":["best", "random"],
              "max_features":[None, 3, 5, 7],    #None= Butun feature' lari kullan
              "max_depth": [None, 4, 5, 6, 7, 8, 9, 10],
              "min_samples_leaf": [1, 2, 3, 5],
              "min_samples_split": [2, 3, 5, 7, 9, 15]}
```

```python
dt_grid_model.best_params_
```
